```python

import pandas as pd

import plotly.express as px

#Carga de datos
df = pd.read_csv('../enaho_2023.csv', encoding='latin-1', sep=None, engine='python')
df.columns = df.columns.str.strip()
df.head()

#Histograma de edades
fig_hist = px.histogram(df, x="edad", title="Prueba de histograma de edades")
fig_hist.show()

#Diagrama de dispersion de ingresos y trabajo
fig_scatter = px.scatter(df, x="horas", y="ingreso", title="Prueba de Dispersión")
fig_scatter.show()
```
